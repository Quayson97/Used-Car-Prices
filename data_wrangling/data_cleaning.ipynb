### Data Cleaning - Used Cars Pricing 

#### Objectives:

 1.  Handle missing values
 2.  Correct data formatting
 3.  Standardize and normalize data

```python
# importing libraries

import pandas as pd
import numpy as np
```

```python
# Link to the data

file_path="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
# cloumn headers for the dataset

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
#Loading the datatset

df=pd.read_csv(file_path, names =headers)
df.head()
```

```python
# Replace missing values with NaN

df.replace("?", np.NaN, inplace=True)
df.head()
```

```python
#Check for Null values
missing_data = df.isnull()
missing_data
```

```python
# Count the number of missing values in each column 

for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

```python
# There are seven columns with NaN values 
df.head()
```

```python
# Replacing the NaN with the mean of the normalized-losses column

avg_norm_loss = df["normalized-losses"].astype('float').mean(axis=0)
df['normalized-losses'].replace(np.NaN, avg_norm_loss, inplace=True)
```

```python
# Replacing the NaN with the mean of the bore column

avg_bore = df["bore"].astype("float").mean()
df["bore"].replace(np.NaN, avg_bore, inplace = True)
```

```python
# Replacing the NaN with the mean of the Stroke column

avg_Stroke = df["stroke"].astype("float").mean()
df["stroke"].replace(np.NaN, avg_Stroke, inplace=True)
```

```python
# Replacing the NaN with the mean of the horsepower column

avg_horsepower = df["horsepower"].astype("float").mean()
df["horsepower"].replace(np.NaN, avg_horsepower, inplace=True)
```

```python
# Replacing the NaN with the mean of the peak-rpm column

avg_rpm = df["peak-rpm"].astype("float").mean()
df["peak-rpm"].replace(np.NaN, avg_rpm, inplace=True)
```

```python
# Replace the NaN with the most occuring value of the number-of-doors column

top_num = df["num-of-doors"].value_counts().idxmax()  # outputs the value that appear the most in the column
df["num-of-doors"].replace(np.NaN, top_num, inplace=True)
```

```python
# drop the whole row for the missing data at the price column

df.dropna(subset=['price'], axis=0, inplace =True)
df.reset_index(drop=True, inplace=False) # reset the index
```

```python
df.head()
```

```python
df.dtypes
```

```python
# correcting the data format of some columns 
df[["bore", "stroke",]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]]  = df[["price"]].astype("float")
df[["peak-rpm"]]  = df[["peak-rpm"]].astype("float")
df.dtypes 
```

### Data Standardization

```python
# Convert fuel consumption from mpg to L/100km
# fromula: L/100km = 235/mpg

df["city-L/100km"] = 235/df["city-mpg"] # For the city-mpg column 
df["highway-L/100km"] = 235/df["highway-mpg"] # for the highway-mpg column
```

```python
df.head()
```

### Data Normalization 

```python
# Normalize the "lenght", "width" and "height" so that their values ranges from 0 to 1

# replace (original value) by (original value)/(maximum value)
df["length"] = df["length"]/df["length"].max() # For length 
df["width"] = df["width"]/df["width"].max()  # For width
df["height"] = df["height"]/df["height"].max()
```

### Binning 

```python

# Create bins from the "horsepower" to categorise them into cars with high-horsepower, medium-horsepower and low-horsepower

# convert "horspower" to correct data type
df["horsepower"] = df["horsepower"].astype(int, copy=True)
```

```python
# Create bins
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
# catgory names 
cat_names = ['Low', 'Medium', 'High']

# Apply the cut() function to determine the categories values of "horsepower" belongs to
df["horsepower-binned"] = pd.cut(df['horsepower'], bins, labels= cat_names, include_lowest=True)
```

```python
# show "horsepower" and "horsepower-binned"
df[["horsepower", "horsepower-binned"]].head(10)
```

```python
# Plot the distribution of bins
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(cat_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# save the clean dataset 
df.to_csv("Cleaned_df.csv")
```
